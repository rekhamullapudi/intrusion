```python
columns = [
"srcip","sport","dstip","dsport","proto","state","dur","sbytes","dbytes",
"sttl","dttl","sloss","dloss","service","sload","dload","spkts","dpkts",
"swin","dwin","stcpb","dtcpb","smeansz","dmeansz","trans_depth","res_bdy_len",
"sjit","djit","stime","ltime","sintpkt","dintpkt","tcprtt","synack","ackdat",
"is_sm_ips_ports","ct_state_ttl","ct_flw_http_mthd","is_ftp_login","ct_ftp_cmd",
"ct_srv_src","ct_srv_dst","ct_dst_ltm","ct_src_ltm","ct_src_dport_ltm",
"ct_dst_sport_ltm","ct_dst_src_ltm","attack_cat","label"
]
```

```python
df = spark.read \
    .option("header", False) \
    .option("inferSchema", True) \
    .csv(f"{PROJECT_ROOT}/data/raw/*.csv")

df = df.toDF(*columns)

print("Rows:", df.count())
print("Columns:", len(df.columns))
df.printSchema()
```

```python
drop_cols = ["srcip","sport","dstip","dsport"]
df = df.drop(*drop_cols)

print("Remaining columns:", len(df.columns))
```

```python
from pyspark.sql.functions import when, col

df = df.withColumn(
    "label_binary",
    when(col("label") == 0, 0).otherwise(1)
)

df.groupBy("label_binary").count().show()
```

```python
df = df.fillna(0)
```

```python
df.write.mode("overwrite").parquet(f"{PROJECT_ROOT}/data/parquet/network_data")
```
